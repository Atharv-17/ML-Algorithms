```python
import pandas as  pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('../DATA/cluster_mpg.csv')
```

```python
df
```

```python
df.describe()
```

```python
df['origin'].value_countscounts()
```

```python
df_w_dummies=pd.get_dummies(df.drop('name', axis=1))
```

```python
df_w_dummies
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()
```

```python
scaled_data=scaler.fit_transform(df_w_dummies)
```

```python
scaled_data
```

```python
scaled_df=pd.DataFrame(scaled_data, columns=df_w_dummies.columns)
```

```python
scaled_df
```

```python
scaled_df.describe()
```

```python
plt.figure(figsize=(15, 8))
sns.heatmap(scaled_df, )
```

```python
sns.clustermap(scaled_df, col_cluster=False)
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
model=AgglomerativeClustering(n_clusters=3)
```

```python
cluster_labels=model.fit_predict(scaled_df)
```

```python
cluster_labels
```

```python
sns.scatterplot(data=df,x='mpg',  y='weight', hue=cluster_labels)
```

```python
model=AgglomerativeClustering(n_clusters=None,distance_threshold=0 )
```

```python
clus_labels=model.fit_predict(scaled_df)
```

```python
clus_labels
```

```python
from  scipy.cluster.hierarchy import dendrogram
from scipy.cluster import hierarchy
```

```python
linkage_matrix=hierarchy.linkage(model.children_)
```

```python
linkage_matrix
```

```python
plt.figure(figsize=(20,10))
dendro = dendrogram(linkage_matrix, truncate_mode='level', p=3)
```
